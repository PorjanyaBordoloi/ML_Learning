```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Generating non-linear data
# these are random datas:
np.random.seed(0) #this is numpy's random number generation
X =  2 - 3 * np.random.normal(0,1,100)
y = X - 2 * (X ** 2) + np.random.normal(-3, 3, 100)

X = X[:, np.newaxis]

# plot the data:
plt.scatter(X,y,color = 'blue')
plt.title('Scatter plot of non-linear data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

#### Now weâ€™ll fit a polynomial regression model of degree 2 to this data.



```python
# Transform the data into polynomial features
poly = PolynomialFeatures(degree = 3)
X_poly = poly.fit_transform(X)

# Fit a linear regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title("Polynomial Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```

### Evaluating the Model

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y,y_pred)
```

```python
print(mse)
print(r2)
```

#### When a model overfits:

It memorizes the training data, capturing not just the underlying pattern but also the noise (random fluctuations).
As a result, it performs exceptionally well on the training data but fails to generalize to new, unseen data (test data). This happens because the model is too complex and tries to fit every little variation in the training data.
Symptoms of Overfitting:
High performance on training data (low MSE, high R-squared).
Poor performance on test data (high MSE, low R-squared) because the model fails to generalize.
